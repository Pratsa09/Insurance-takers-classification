```python
insurance_train['CallStart']=pd.to_numeric(insurance_train['CallStart'],errors='coerce')
#sinc CallStart is an object datatype
```

```python
insurance_train['CallStart_Hour']=(insurance_train['CallStart']/10000).astype(int)
#use type to get the rounded off integer value
```

```python
insurance_train['CallStart_minute']=((insurance_train['CallStart']%10000)/100).astype(int)
insurance_train['CallStart_Second']=((insurance_train['CallStart']%10000)%100).astype(int)
```

```python
insurance_train.head()
```

```python
#the same process for CallEnd
insurance_train['CallEnd']=insurance_train['CallEnd'].str.replace(':','')
insurance_train['CallEnd']=pd.to_numeric(insurance_train['CallEnd'],errors='coerce')
insurance_train.dtypes
```

```python
insurance_train['CallEnd_Hour']=(insurance_train['CallEnd']/10000).astype(int)
insurance_train['CallEnd_minute']=((insurance_train['CallEnd']%10000)/100).astype(int)
insurance_train['CallEnd_second']=((insurance_train['CallEnd']%10000)%100).astype(int)
```

```python
insurance_train.head()
```

```python
insurance_train['total_start_seconds']=(insurance_train['CallStart_Hour']*3600)+(insurance_train['CallStart_minute']*60)+(insurance_train['CallStart_Second'])
insurance_train['total_End_seconds']=(insurance_train['CallEnd_Hour']*3600)+(insurance_train['CallEnd_minute']*60)+(insurance_train['CallEnd_second'])
```

```python
insurance_train.head()
```

```python
insurance_train['call_duration']=(insurance_train['total_End_seconds'])-(insurance_train['total_start_seconds'])
```

```python
insurance_train=insurance_train.drop(['CallStart','CallEnd','CallStart_Hour','CallStart_minute','CallStart_Second','CallEnd_Hour','CallEnd_minute','CallEnd_second','total_start_seconds','total_End_seconds'],axis=1)
```

```python
x=insurance_train.drop('CarInsurance',axis=1)
y=insurance_train['CarInsurance']
from sklearn.model_selection import train_test_split
#traintest split
x_train, x_validation, y_train, y_validation = train_test_split(x,y, train_size=0.7, random_state=100)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
#the above information are called Hyperparameters.
```
